```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import gridspec
import seaborn as sns

plt.rcParams["text.usetex"] = True
plt.rcParams["font.family"] = "serif"

sns.set_palette('Set1',color_codes=True)

DATDIR = './dat/'
PLTDIR = './plt/'
```

```python
# SKI5_2009 population

eos = 'SKI5'

short_scenario = '3G'
scenario = '3Gm1'
scenario_dict = {'O5': 'A+ (1 yr)', 'O5y2': 'A+ (2 yr)', '3Gd1': 'XG (1 d)', '3Gw1': 'XG (1 wk)', '3Gm1': 'XG (1 mo)'}
color = sns.color_palette()[1]
pop = 'v0'

poppath = DATDIR+'unif_BNS_{1}{2}/{0}_unif_BNS_{1}_detected-{3}.csv'.format(short_scenario,eos,pop,scenario)
pop_dat = np.genfromtxt(poppath,names=True,delimiter=',',dtype=None,encoding=None)
mcs = pop_dat['mc']
Lts = pop_dat['Lt']
mcs_det = pop_dat['mc_det']
Lts_det = pop_dat['Lt_det']
dLts = pop_dat['dLt']

plt.figure(figsize=(6,4.5))
plt.scatter(mcs,Lts,marker='+',c=color,label='SKI5, injected',alpha=0.5,zorder=0)

eos = 'SKI52009'

short_scenario = '3G'
scenario = '3Gm1'
scenario_dict = {'O5': 'A+ (1 yr)', 'O5y2': 'A+ (2 yr)', '3Gd1': 'XG (1 d)', '3Gw1': 'XG (1 wk)', '3Gm1': 'XG (1 mo)'}
color = sns.color_palette()[2]
pop = 'v0'

poppath = DATDIR+'unif_BNS_{1}{2}/{0}_unif_BNS_{1}_detected-{3}.csv'.format(short_scenario,eos,pop,scenario)
pop_dat = np.genfromtxt(poppath,names=True,delimiter=',',dtype=None,encoding=None)
mcs = pop_dat['mc']
Lts = pop_dat['Lt']
mcs_det = pop_dat['mc_det']
Lts_det = pop_dat['Lt_det']
dLts = pop_dat['dLt']

plt.scatter(mcs,Lts,marker='+',c=color,label='SKI5_2009, injected',alpha=0.8,zorder=1)
plt.scatter(mcs_det,Lts_det,marker='.',alpha=0.2,c='k',s=4*20.*dLts/np.mean(dLts),label='SKI5_2009, XG (1\,m)',zorder=-1)

plt.yscale('log')
plt.xlim(0.8,2.)
plt.ylim(4.,8000.)
plt.ylabel(r'$\tilde{\Lambda}$',size=18)
plt.xlabel(r'$\mathcal{M}\;[M_\odot]$',size=18)

plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.legend(frameon=False)
plt.savefig(PLTDIR+'SKI52009_pop.png')
```
